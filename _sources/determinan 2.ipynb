MAKA :                   
$C21$     =     $(-1)^2+^1 Det(M21)$                                
$C22$     =     $(-1)^2+^2 Det(M22)$                        
$C23$     =     $(-1)^2+^3 Det(M23)$                 
$C24$     =     $(-1)^2+^4 Det(M24)$                          

## 1. Mencari $DetM21$    
$                       
M21 = \begin{bmatrix}
1 & 2 & 3 \\
3 & 1 & 1 \\
1 & 2 & 1\\
\end{bmatrix}
$                       

  C21   =  $(-1)^2+^1 Det(M21)$                                      
  C22   =  $(1)^2+^2 Det(M22)$                                         
  C23   =  $(-1)^2+^3 Det(M23)$ 

  A. mencari det(m21) dari matrix M21                               
      $
      M21C21 = \begin{bmatrix}
      2 & 3 \\
      2 & 1\\                                                  
      \end{bmatrix}
      $
      determinannya M21 adalah -4                           
                                   
  B. mencari det(M22) dari matrix M21                               
      $
      M21C22 = \begin{bmatrix}
      1 & 3 \\
      1 & 1\\                                                  
      \end{bmatrix}
      $
      determinannya M22 adalah -2 
    
  C.  mencari det(M23) dari matrix M21                               
      $
      M21C22 = \begin{bmatrix}
      1 & 2 \\
      1 & 2\\                                                  
      \end{bmatrix}
      $
      determinannya M23 adalah 0

  jadi :                              
  C21   =  $3 * (-1)^2+^1 -4$   =     12                                            
  C22   =  $1 * (1)^2+^2 -2$  =     -2                                                     
  C23   =  $1 * (-1)^2+^3 0$  =     0                                   

  $DetM31$    = 12 + -2 + 0 = 10
  

  ## 2. Mencari $DetM22$    
$                       
M22 = \begin{bmatrix}
1 & 2 & 3 \\
2 & 1 & 1 \\
2 & 2 & 1\\
\end{bmatrix}
$                       

  C21   =  $(-1)^2+^1 Det(M21)$                                      
  C22   =  $(1)^2+^2 Det(M22)$                                         
  C23   =  $(-1)^2+^3 Det(M23)$ 


  A. mencari det(m21) dari matrix M22                              
      $
      M32C21 = \begin{bmatrix}
      2 & 3 \\
      2 & 1\\                                                  
      \end{bmatrix}
      $
      determinannya M21 adalah -4                           
                                   
  B. mencari det(m22) dari matrix M32                              
      $
      M32C22 = \begin{bmatrix}
      1 & 3 \\
      2 & 1\\                                                  
      \end{bmatrix}
      $
      determinannya M22 adalah -5

  C. mencari det(m23) dari matrix M22                               
      $
      M32C22 = \begin{bmatrix}
      1 & 2 \\
      2 & 2\\                                                  
      \end{bmatrix}
      $
      determinannya M22 adalah -2


  jadi :                              
  C21   =  $2 * (-1)^2+^1 -4$   =    8                                           
  C22   =  $1 * (1)^2+^2 -5$  =      -5                                                  
  C23   =  $1 * (-1)^2+^3 -2$  =     2                                  

  $DetM32$    = 8 + (-5) + 2 = 5


## 3. Mencari $DetM23$    
$                       
M23 = \begin{bmatrix}
1 & 1 & 3 \\
2 & 3 & 1 \\
2 & 1 & 1\\
\end{bmatrix}
$                       

  C21   =  $(-1)^2+^1 Det(M21)$                                      
  C22   =  $(1)^2+^2 Det(M22)$

      M33C22 = \begin{bmatrix}
      1 & 3 \\

      \end{bmatrix}
      $

  C. mencari det(m23) dari matrix M23

      1 & 1 \\

      \end{bmatrix}

  jadi :                              
  C21   =  $2 * (-1)^2+^1 -2$ =    4                                          
  C22   =  $3 * (1)^2+^2 -5$  =    -15                                                  
  C23   =  $1 * (-1)^2+^3 -1$  =    1                                  

  $DetM33$    = 4 + -15 + 1 = 10
